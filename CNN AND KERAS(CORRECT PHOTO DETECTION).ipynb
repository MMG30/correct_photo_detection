```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import GlobalAveragePooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
#Initialize CNN
classifier=Sequential()
```

```python
#Add 1 convolution layers
classifier.add(Conv2D(32,(5,5),input_shape =(28,28,3),activation='relu'))
```

```python
#Add pooling layer
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#Add 1 more convolution layers
classifier.add(Conv2D(64,(5,5),input_shape=(28,28,3),activation='relu'))
```

```python
#Add 1 more pooling layer
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#Add global average pooling layer
classifier.add(Flatten())
```

```python
#Add full connection
classifier.add(Dense(units=128,activation='relu'))
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
#Compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fit CNN to images
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)

test_datagen=ImageDataGenerator(rescale=1./255)

train_set=train_datagen.flow_from_directory('C://Users/mayan/Desktop/Machine Learning/dataset/training_set',
                                            target_size=(28,28),batch_size=32,class_mode='binary')
test_set=test_datagen.flow_from_directory('C://Users\mayan/Desktop/Machine Learning/dataset/test_set',
                                          target_size=(28,28),batch_size=32,class_mode='binary')
```

```python
classifier.fit_generator(train_set,steps_per_epoch=15,epochs=20,validation_data=test_set,validation_steps=20)

classifier.save('model_binary_complex.h5')
```
